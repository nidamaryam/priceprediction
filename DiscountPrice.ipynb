```python
os.chdir('C:\\Users\\Nida Maryam\\Documents\\SummerAnalytics')
df=pd.read_csv('train_final1.csv')
df.head()
```

```python
print(df.shape)
```

```python
df.rename(columns={'Offer %':'offper'}, inplace=True)
df.columns
```

```python
print(df.isnull().sum())
```

```python
df.loc[df['title'].str.contains('Women'), 'maincateg'] = 'Women'
df.loc[df['title'].str.contains('Men'), 'maincateg'] = 'Men'
df['maincateg'].fillna(value='Unisex', inplace=True)
df.isnull().sum()
```

```python
df.offper=df.offper.str.replace('%', '')
df['offper'] = df.offper.astype(float)
df=df.drop(['id', 'title'],axis=1)
df.dtypes
```

```python
cat_cols=['platform','maincateg','fulfilled1']

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df[cat_cols] = df[cat_cols].apply(lambda col: le.fit_transform(col))

df.head(20)
```

```python

from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5)
knn_ar = imputer.fit_transform(df)

df = pd. DataFrame(knn_ar, columns=['Rating', 'maincateg', 'platform', 'price1','actprice1',
       'offper', 'norating1', 'noreviews1', 'star_5f', 'star_4f', 'star_3f',
       'star_2f', 'star_1f', 'fulfilled1']) 
df.head(20)
```

```python
df['maincateg'] = df.maincateg.astype(int)
df['platform'] = df.platform.astype(int)
df['fulfilled1'] = df.fulfilled1.astype(int)
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(['offper','price1'],axis=1)
y = df['price1']
X_train, X_test, y_train ,y_test = train_test_split( X, y, test_size=0.5,  random_state=0 ) 
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV

RMSE_test = []
RMSE_train = []
Model_Name = []
```

```python
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import ElasticNet
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.linear_model import Lasso
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
```

```python
model_list = [RandomForestRegressor(),XGBRegressor(),DecisionTreeRegressor(),ElasticNet(), # SVR(),
              GradientBoostingRegressor(),ExtraTreesRegressor(),
              AdaBoostRegressor(),Lasso(),LinearRegression(),Ridge()]

for model in model_list:
    print(model)
    model.fit(X_train, y_train)

    y_predictions = model.predict(X_test)
    mse = mean_squared_error(y_test, y_predictions)
    rmse = np.sqrt(mse)
    score = model.score(X_test,y_test)

    RMSE_test.append(rmse)
    Model_Name.append( 'RandomForestRegressor' )

    print('Root Mean Square Error : ', rmse )
    print('R2 Score : ', score )
    print('\n')
```
